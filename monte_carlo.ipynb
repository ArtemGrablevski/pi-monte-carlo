# Estimating the value of Pi using Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dots_total = 20_000

x = np.random.uniform(-1, 1, dots_total)
y = np.random.uniform(-1, 1, dots_total)

inside_circle = x ** 2 + y ** 2 < 1

dots_inside = np.sum(inside_circle)

pi = 4 * dots_inside / dots_total

print("Approximated π:", pi)
```

```python
plt.figure(figsize=(7, 7))

# red dots outside the circle
plt.scatter(x=x[~inside_circle], y=y[~inside_circle], color="red", s=1)
# blue dots inside the circle
plt.scatter(x=x[inside_circle], y=y[inside_circle], color="blue", s=1)

# coordinate axes
plt.axhline(y=0, color="black", linewidth=1)
plt.axvline(x=0, color="black", linewidth=1)

# axis arrows
plt.annotate("", xy=(0, 1), xytext=(0, -1), arrowprops=dict(arrowstyle="->", lw=1, color="black"))
plt.annotate("", xy=(1, 0), xytext=(-1, 0), arrowprops=dict(arrowstyle="->", lw=1, color="black"))

# black circle
circle = plt.Circle((0, 0), 1, color="black", fill=False, linewidth=1)
plt.gca().add_patch(circle)

plt.title("Estimating the value of π using Monte Carlo", pad=15)

plt.text(
    0.5, -1.2, f"Approximated π: {pi}", ha="center", fontsize=12, 
    bbox=dict(facecolor="white", edgecolor="black", boxstyle="round,pad=0.4")
)

plt.axis("equal")

plt.xlim(-1, 1)
plt.ylim(-1, 1)

plt.grid(True)

plt.show()
```
